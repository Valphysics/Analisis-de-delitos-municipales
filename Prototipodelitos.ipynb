```python
#creación de un DataFrame con los datos a utilizar
import pandas as pd
query = "SELECT Año, Entidad, Municipio, Tipo_de_delito, Subtipo_de_delito, Enero, Febrero, Marzo, Abril, Mayo, Junio, Julio, Agosto, Septiembre, Octubre, Noviembre, Diciembre FROM Municipal_Delitos2015_2021"
df = pd.read_sql(query, conn)
conn.close() #cerrar la conexión después de usar pandas
print(df.head())
```

```python
#Modificacion de los meses a formato largo
df_largo = df.melt(id_vars=["Año","Entidad", "Municipio", "Tipo_de_delito", "Subtipo_de_delito"], 
                   value_vars=["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio",
                               "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"],
                   var_name="Mes",
                   value_name="Cantidad_Delitos")

df_largo.sample(10) 
```

```python
# Agrupar y obtener el promedio
promedio = df_largo.groupby(["Año", "Municipio", "Mes"])["Cantidad_Delitos"].mean().reset_index()
promedio
```

El promedio es bajo, ya que tenemos muchos valores en 0 /
Ahora vamos a hacer un analisis solo con los Delitos reportados

```python
df_filtrado = df_largo[df_largo["Cantidad_Delitos"] > 0]

# Suma total por municipio
top_municipios = df_filtrado.groupby(["Año","Municipio","Mes"])["Cantidad_Delitos"].sum().sort_values(ascending=False).reset_index()
top_municipios.head(10)
```

Ahora veremos el delito mas comun y la temporada con mayor numero de delitos

```python
delito_mascomun= df_filtrado.groupby(["Municipio", "Tipo_de_delito"])["Cantidad_Delitos"].sum().sort_values(ascending=False).reset_index()
delito_mascomun.head(10)
```

```python
temporada_topcrimenes= df_filtrado.groupby(["Año", "Mes"])["Cantidad_Delitos"].sum().sort_values(ascending=False).reset_index()
temporada_topcrimenes.head(10)
```

promedio de delitos en ciudad de mexico

```python
df_cdmx = df_largo[df_largo["Entidad"] == "Ciudad de México"]

prom_delitos = df_cdmx.groupby(["Año", "Mes", "Tipo_de_delito"])["Cantidad_Delitos"].mean().sort_values(ascending=False).reset_index()
prom_delitos
```

```python
#Mayor cantidad de homicidios
Homicidios = df_largo[df_largo["Tipo_de_delito"] == "Homicidio"]
homicidios_count = Homicidios.groupby(["Año", "Municipio"])["Cantidad_Delitos"].sum().sort_values(ascending=False).reset_index()
homicidios_count.head(10)
```

```python
#mayor cantidad de feminicidios
feminicidios = df_largo[df_largo["Tipo_de_delito"] == "Feminicidio"]
feminicidios_count = feminicidios.groupby(["Año", "Municipio"])["Cantidad_Delitos"].sum().sort_values(ascending=False).reset_index()
feminicidios_count.head(10)
```

Analisis de riesgo de delitos a vehiculos privados por municipio

```python
# Total de delitos por municipio y año
total_delitos = df_largo.groupby(["Año", "Municipio"])["Cantidad_Delitos"].sum().reset_index()
total_delitos.rename(columns={"Cantidad_Delitos": "Delitos_Totales"}, inplace=True)

delitos_auto = df_largo[df_largo["Subtipo_de_delito"].str.contains("vehículo|autopartes", case=False, na=False)]
delitos_auto_municipio = delitos_auto.groupby(["Año", "Municipio"])["Cantidad_Delitos"].sum().sort_values(ascending=False).reset_index()
delitos_auto_municipio.rename(columns={"Cantidad_Delitos": "Delitos_Automovil"}, inplace=True)
delitos_auto_municipio.head(10)
```

```python
df_riesgo = pd.merge(total_delitos, delitos_auto_municipio, on=["Año", "Municipio"], how="left")
df_riesgo["Porcentaje_Delitos_Auto"] = (df_riesgo["Delitos_Automovil"] / df_riesgo["Delitos_Totales"]) * 100
df_riesgo = df_riesgo.fillna(0).sort_values(ascending=False, by='Delitos_Totales') # Rellenar NaN con 0
df_riesgo.head(30)
```
